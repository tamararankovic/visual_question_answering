```python
from transformers import BertTokenizer, BertModel
import pandas as pd
```

```python
def fv2str(fv):
    fvstr = ''
    last = len(fv)-1
    for iv, v in enumerate(fv):
        fvstr += str(v)
        if not last == iv:
            fvstr += '|'
    return fvstr

def str2fv(string):
    fv = []
    for v in string.split('|'):
        fv.append(float(v))
    return fv
```

```python
images = []
questions = []

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

for i, question in enumerate(train.question.values):
    inputs = tokenizer(question, return_tensors='pt')
    outputs = model(**inputs)
    cls_embedding = outputs.last_hidden_state[0][0].tolist()
    questions.append(fv2str(cls_embedding))
    images.append(train.image.values[i])
    print(train.image.values[i])
    
df = pd.DataFrame()
df['image'] = images
df['question'] = questions
df.head()
df.to_csv('data/train_questions.csv')
```
